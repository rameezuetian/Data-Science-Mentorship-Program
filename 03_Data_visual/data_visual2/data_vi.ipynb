```python
# scatter plot
# between tow numerical data 
# catogorical data for hue  
# scatter size and figure(width and height) 
plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x='age',y='fare', hue='sex')
```

# how to interpret 
1. we have Age of passengers at x-axis and fare at y-axis
2. The sactter plot shows the relationship between age and fare of titanic passenegers.
3. we further grouped the passenger by their genders such as male (green and red).
4. Two male passengers gave the highest fare in titanic, and they might be travelling cot 
5. The oldest person at titanic is a;so a male having 80 years of age shown in scatter plot.(As shown in fig.)



```python
# convert the same plot into  box plot 
#style in graph
#sns.set_theme(style='darkgrid',font_scale=1.5, font=)
pallte = {'First':'green','Second':'red','Third':'blue'}
plt.figure(figsize=(8,4))
#sns.boxenplot(data=df, x='sex',y='age', hue='class',palette=pallte)
sns.boxplot(data=df, x='sex',y='age', hue='class',palette=pallte)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box Plot of Age  based on  Gender')


# change legend position and titile
plt.legend(loc='upper right', title='Class Name')

# change the x tick labels/names
plt.xticks(ticks=[0,1] ,labels=["Males","Female"])

# rotate the x ticks
plt.xticks(rotation=40)
plt.yticks(rotation=50)
```

```python
#style in graph
sns.set_theme(style='darkgrid',font_scale=1.5, font='Times New Roman')
pallte = {'male':'green','female':'red'}
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x='age',y='fare', hue='sex', size='fare',palette=pallte)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
```

# bar plot usinf seaborn

```python
# convert the same plot into  box plot 
#style in graph
sns.set_theme(style='darkgrid',font_scale=1.5, font='Times New Roman')
pallte = {'First':'green','Second':'red','Third':'blue'}
plt.figure(figsize=(8,4))
#sns.boxenplot(data=df, x='sex',y='age', hue='class',palette=pallte)
sns.barplot(data=df, x='sex',y='age', hue='class',palette=pallte, capsize=0.1)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box Plot of Age  based on  Gender')


# change legend position and titile
plt.legend(loc='upper right', title='Class Name')

# change the x tick labels/names
plt.xticks(ticks=[0,1] ,labels=["Males","Female"])

# rotate the x ticks
plt.xticks(rotation=40)
plt.yticks(rotation=50)


# agar x per vategorical ah y per numerica ha to hum uska bar aur boxplot bhi bana saktai ha 



# add the bar error
```
