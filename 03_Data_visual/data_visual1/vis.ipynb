# Data Viz 

```python
import pandas as pd
import numpy as np 
import  matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = sns.load_dataset('titanic')
```

```python
df.plot(kind = 'scatter', x='age' , y = 'fare' , color='#190ee8')
# color blinding 
```

```python
df.groupby('sex')['age'].mean().plot(kind='barh')
plt.ylabel('chal iyee')
plt.xlabel('sahi ha ')
```

```python
# seaborn 

# sns.scatterplot(data=df, x = 'age', y = 'fare', hue='class')
sns.scatterplot(data=df, x = 'age', y = 'fare', hue='sex')
```
