```python
import pandas as pd 
df = pd.DataFrame({'name':['John Author Doe', 'Jane Ann Smith','Sajjad Ali', 'Abu Bakr'],
                   'Location':['Los Angeles , Ca ','Wshington , DC ','Lahire, Pakistan',' Multan, Pakistna']})
df 
```

```python
df.name.str.split(' ',expand=True)
```

```python
df[["FirtName"]] = df.name.str.split(' ,',expand=True)
df 
```

```python
df[["City_Name","Country_Name"]] = df.Location.str.split(',',expand=True)
df 
```
